##  <b> Importar librerГ­as y cargar datos

```python
import pandas as pd
datos = pd.read_csv("celsius.csv")
```

##  <b>Mostrar las primeras filas

```python
datos.head()
```

```python
datos["Celsius"].values
```

## <b> Separar entradas y salidas

```python
x = datos["Celsius"].values
y = datos["Fahrenheit"].values
```

## <b> Crear el modelo neuronal

```python
import tensorflow as tf

modelo = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)), #la entrada, los clsius: 1 solo dato
    tf.keras.layers.Dense(units=1),#la salid. Tambien s un dato: Los grados fahrenheit
])

modelo.compile(
    optimizer=tf.keras.optimizers.Adam(.1),
    loss="mean_squared_error"
)
```

## <b> Entrenamiento del modelo

```python
entrenamiento = modelo.fit(x, y, epochs=1000)
```

## <b> Graficar la pГ©rdida

```python
import matplotlib.pyplot as plt

plt.xlabel("# Epoca")
plt.ylabel("Magnitud de la perdida")
plt.plot(entrenamiento.history["loss"])
```

## <b> Realizar una predicciГіn

```python
import tensorflow as tf

c = 1230
prediccion = modelo.predict(tf.constant([c]))
print(f"Prediccion: {c} celsius son {prediccion[0][0]} fahrenheit")
```
